I am a bit of a music snob. I listen to Spotify every day, whether I am working at home or driving around in my car. I have attempted to change my main music listening app multiple times, to Tidal or Rdio, etc. However, Spotify has a feature called "Discover Weekly" that suggests songs for me based on my listening habits. This "Discover Weekly" playlist has truly blown me away! (If you have never tried it, use Spotify for a couple of weeks, then search for the "Discover Weekly" playlist; I think you will agree!)

Since the songs in my Discover Weekly playlist are only available to me for one week, last year I started throwing them into a playlist called "Discover Weekly History". I don't remember to do this every week, but I do it at least once per month. This playlist keeps my songs around for a longer duration than the Discover Weekly playlist does. 

For this project, I decided to open my Discover Weekly History playlist, and to graph my songs to see if I had any changes to my listening habits over the year that I had been tracking them. I also logged in to the website Spotify.me, which created a playlist for me based on my overall listening history. Finally, I created a playlist of songs that I actively disliked, first by adding songs that I knew were bad (come on, who actually *likes* Nickelback?), and then by asking a good friend whose music I do not like AT ALL to please send me his own playlist from Spotify.me, and adding artists and songs from the result.

I graphed each of the audio features of my Discover Weekly History playlist over time to see if I could see any trends emerging. I also compared the audio features of my "good songs" playlist to the audio features of the "bad songs" playlist, to see if there was any feature that I had a strong preference for or dislike of. Now I am going to combine the songs into one database and train models to predict which songs I will like.

```python
thebad['target']= 0
thegood['target']=1

thegood
```

```python
allsongs = thebad.append(thegood)
allsongs
```

```python
#Here I am selecting the things that I would like to look at.
features = ["danceability", "loudness", "valence", "energy", "instrumentalness", "acousticness", "key", "speechiness","duration_ms"]
```

```python
train, test  =  train_test_split(allsongs, test_size = 0.15)
print("Training size: {}, Test size: {}".format(len(train),len(test)))
```

```python
#Split the data into x and y test and train sets to feed them into a bunch of classifiers!
x_train = train[features]
y_train = train["target"]

x_test = test[features]
y_test = test["target"]
```

```python
c = DecisionTreeClassifier(min_samples_split=100)
dt = c.fit(x_train, y_train)
```

```python
def show_tree(InputTree, features, path):
    f = io.StringIO()
    tree.export_graphviz(InputTree, out_file=f, feature_names=features)
    pydotplus.graph_from_dot_data(f.getvalue()).write_png(path)
    img = misc.imread(path)
    plt.rcParams["figure.figsize"] = (20,20)
    plt.imshow(img)
show_tree(dt, features, "dec_tree.png")
```

This decision tree is hard to read but it shows the best steps to determin if I like a song or not. Then I begin the models! I ran many. The accuracy of each model is listed below it.

```python
y_pred = c.predict(x_test)
score = accuracy_score(y_test, y_pred) * 100
print("Accuracy using Decision Tree: ", round(score, 1), "%")
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(3)
knn.fit(x_train, y_train)
knn_pred = c.predict(x_test)
score = accuracy_score(y_test, knn_pred) * 100
print("Accuracy using Knn Tree: ", round(score, 1), "%")
```

```python
from sklearn.neural_network import MLPClassifier
mlp = MLPClassifier()
mlp.fit(x_train, y_train)
mlp_pred = mlp.predict(x_test)
score = accuracy_score(y_test, mlp_pred) * 100
print("Accuracy using mlp Tree: ", round(score, 1), "%")
```

```python
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier(max_depth=5, n_estimators=10, max_features=1)
forest.fit(x_train, y_train)
forest_pred = forest.predict(x_test)
score = accuracy_score(y_test, forest_pred) * 100
print("Accuracy using random forest: ", round(score, 1), "%")
```

```python
from sklearn.ensemble import AdaBoostClassifier
ada = AdaBoostClassifier(n_estimators=100)
ada.fit(x_train, y_train)
ada_pred = ada.predict(x_test)
score = accuracy_score(y_test, ada_pred) * 100
print("Accuracy using ada: ", round(score, 1), "%")
```

```python
from sklearn.naive_bayes import GaussianNB
gauss = GaussianNB()
gauss.fit(x_train, y_train)
gauss_pred = gauss.predict(x_test)
score = accuracy_score(y_test, gauss_pred)*100
print("Accuracy using gauss: ", round(score, 1), "%")
```

```python
from sklearn.cluster import KMeans
k_means = KMeans(n_clusters=3, random_state=0)
k_means.fit(x_train, y_train)
predicted= k_means.predict(x_test)
score = accuracy_score(y_test, predicted)*100
print("Accuracy using Kmeans: ", round(score, 1), "%")
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gbc = GradientBoostingClassifier(n_estimators=100, learning_rate=.1, max_depth=1, random_state=0)
gbc.fit(x_train, y_train)
predicted = gbc.predict(x_test)
score = accuracy_score(y_test, predicted)*100
print("Accuracy using Gbc: ", round(score, 1), "%")
```

```python
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
qda = QuadraticDiscriminantAnalysis()
qda.fit(x_train, y_train)
qda_pred = qda.predict(x_test)
score = accuracy_score(y_test, qda_pred)*100
print("Accuracy using qda: ", round(score, 1), "%")
```

```python
from sklearn.svm import SVC# from s 
svc_lin = SVC(kernel="linear", C=0.025)
svc_lin.fit(x_train, y_train)
svc_pred = svc_lin.predict(x_test)
score = accuracy_score(y_test, svc_pred) * 100
print("Accuracy using svc linear: ", round(score, 1), "%")
```

```python
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.gaussian_process.kernels import RBF
gpc = GaussianProcessClassifier(1.0 * RBF(1.0), warm_start=True)
gpc.fit(x_train, y_train)
gpc_pred = gpc.predict(x_test)
score = accuracy_score(y_test, gpc_pred) * 100
print("Accuracy using gpc: ", round(score, 1), "%")
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(x_train, y_train)
score = clf.score(x_test, y_test) * 100
print("Accuracy using clf: ",round(score, 1), "%")
```
