```python
import matplotlib.pyplot as plt
import mpld3
#mpld3.enable_notebook()
from mpld3 import plugins
from iex import stock
```

```python
A = stock('A').chart_table(range='1y')
A = A.rename(columns={'close':'A'})
AAPL = stock('AAPL').chart_table(range='1y')
AAPL = AAPL.rename(columns={'close':'AAPL'})
AMD = stock('AMD').chart_table(range='1y')
AMD = AMD.rename(columns={'close':'AMD'})
ARQL = stock('ARQL').chart_table(range='1y')
ARQL = ARQL.rename(columns={'close':'ARQL'})
AU = stock('AU').chart_table(range='1y')
AU = AU.rename(columns={'close':'AU'})
BHP = stock('BHP').chart_table(range='1y')
BHP = BHP.rename(columns={'close':'BHP'})
BLIN  = stock('BLIN').chart_table(range='1y')
BLIN = BLIN.rename(columns={'close':'BLIN'})
BOSC = stock('BOSC').chart_table(range='1y')
BOSC = BOSC.rename(columns={'close':'BOSC'})
BP = stock('BP').chart_table(range='1y')
BP = BP.rename(columns={'close':'BP'})
BTI= stock('BTI').chart_table(range='1y')
BTI = BTI.rename(columns={'close':'BTI'})
```

```python
df_1y_date = df_1y['date'].values.tolist()
df_1y_a = df_1y['A'].values.tolist()
```
